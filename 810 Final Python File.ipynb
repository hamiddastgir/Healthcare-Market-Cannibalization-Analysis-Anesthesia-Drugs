```python
##  Aggregating Claims Data by Year and Procedure Code ##

import matplotlib.pyplot as plt
import numpy as np

# Filter data for the years 2016 to 2018
claims_filtered_years = final_df[final_df['Year'].isin([2016, 2017, 2018])]

# Group by year and HCPCS code (product) to count claims, patients, and HCPs
claims_grouped = claims_filtered_years.groupby(['Year', 'Procedure_Code']).agg(
    total_claims=('Claim_ID', 'count'),
    unique_patients=('Patient_ID', pd.Series.nunique),
    unique_hcps=('HCP_ID', pd.Series.nunique)
).reset_index()

# Calculate total claims per year for percentage calculation
total_claims_per_year = claims_grouped.groupby('Year')['total_claims'].sum()

```

```python
# Calculate share of claims, patients, and HCPs per product per year
claims_grouped['claims_percentage'] = claims_grouped.apply(lambda x: (x['total_claims'] / total_claims_per_year[x['Year']]) * 100, axis=1)
claims_grouped['patients_percentage'] = claims_grouped.apply(lambda x: (x['unique_patients'] / claims_grouped[claims_grouped['Year'] == x['Year']]['unique_patients'].sum()) * 100, axis=1)
claims_grouped['hcps_percentage'] = claims_grouped.apply(lambda x: (x['unique_hcps'] / claims_grouped[claims_grouped['Year'] == x['Year']]['unique_hcps'].sum()) * 100, axis=1)

# Pivot data for stacked bar charts
claims_percentage_pivot = claims_grouped.pivot(index='Year', columns='Procedure_Code', values='claims_percentage')
patients_percentage_pivot = claims_grouped.pivot(index='Year', columns='Procedure_Code', values='patients_percentage')
hcps_percentage_pivot = claims_grouped.pivot(index='Year', columns='Procedure_Code', values='hcps_percentage')
```

```python
# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(13, 15), sharex=True)

# Function to add labels to bars
def add_labels(ax, data):
    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f%%', label_type='center', color='white')

# Defining Function fo Bar chart
def stacked_bar_chart(ax, data, title, ylabel):
    data_perc = data.divide(data.sum(axis=1), axis=0) * 100
    data_perc.plot(kind='bar', stacked=True, ax=ax)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    add_labels(ax, data_perc)  # Add labels with percentage format

# Plot claims percentage
stacked_bar_chart(axes[0], claims_percentage_pivot, 'Trend of Claims Percentage for the Products', 'Claims Percentage')

# Plot unique patients
stacked_bar_chart(axes[1], patients_percentage_pivot, 'Trend of Unique Patients for the Products', 'Unique Patients Percentage')

# Plot unique HCPs
stacked_bar_chart(axes[2], hcps_percentage_pivot, 'Trend of Unique HCPs for Products', 'Unique HCPs Percentage')

# Add legend and adjust layout
plt.legend(title='Procedure Code/Products', loc='upper right')
plt.tight_layout()
```

```python
claims_percentage_pivot.to_csv('claims_percentage_pivot.csv', index=True)
patients_percentage_pivot.to_csv('patients_percentage_pivot.csv', index=True)
hcps_percentage_pivot.to_csv('hcps_percentage_pivot.csv', index=True)

fig.savefig('stacked_bar_charts.png')
```

```python
##  Average Claims per HCP by Year and Procedure Code ##

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Filter data for the years 2016 to 2018 and the specified procedure codes
filtered_df = final_df[
    final_df['Year'].isin([2016, 2017, 2018]) &
    final_df['Procedure_Code'].isin(['J1885', 'J2250', 'J2704', 'J3010'])
]

# Group and calculate for both Patients and Claims
grouped_data = filtered_df.groupby(['Year', 'Procedure_Code'])
patients_per_year_product = grouped_data['Patient_ID'].nunique().reset_index()
claims_per_year_product = grouped_data['Claim_ID'].count().reset_index()
hcps_per_year_product = grouped_data['HCP_ID'].nunique().reset_index()
```

```python
# Merge the DataFrames on 'Year' and 'Procedure_Code'
merged_patients = pd.merge(patients_per_year_product, hcps_per_year_product, on=['Year', 'Procedure_Code'])
merged_claims = pd.merge(claims_per_year_product, hcps_per_year_product, on=['Year', 'Procedure_Code'])

# Calculate patients and claims per writer (HCP)
merged_patients['Patients_per_HCP'] = merged_patients['Patient_ID'] / merged_patients['HCP_ID']
merged_claims['Claims_per_HCP'] = merged_claims['Claim_ID'] / merged_claims['HCP_ID']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for the years 2016 to 2018 and the specified procedure codes
filtered_df = final_df[
    final_df['Year'].isin([2016, 2017, 2018]) &
    final_df['Procedure_Code'].isin(['J1885', 'J2250', 'J2704', 'J3010'])
]

# Get the first year each HCP prescribed each brand
first_prescription_year = filtered_df.groupby(['HCP_ID', 'Procedure_Code'])['Year'].min().reset_index()

# Filter for new writers (first prescription in the specified years)
new_writers = first_prescription_year[first_prescription_year['Year'].isin([2016, 2017, 2018])]

# Group by year and procedure code to count new writers
new_writers_count = new_writers.groupby(['Year', 'Procedure_Code'])['HCP_ID'].count().reset_index(name='New_Writer_Count')

# Group by year and procedure code to count total writers
total_writers_count = filtered_df.groupby(['Year', 'Procedure_Code'])['HCP_ID'].nunique().reset_index(name='Total_Writer_Count')

# Merge the DataFrames to calculate continuing writers
continuing_writers_count = pd.merge(total_writers_count, new_writers_count, on=['Year', 'Procedure_Code'], how='left').fillna(0)
continuing_writers_count['Continuing_Writer_Count'] = continuing_writers_count['Total_Writer_Count'] - continuing_writers_count['New_Writer_Count']

# Filter for the years 2017 and 2018
continuing_writers_count = continuing_writers_count[continuing_writers_count['Year'].isin([2017, 2018])]

# Create a line chart using seaborn with data points
plt.figure(figsize=(10, 6))
ax = sns.lineplot(x='Year', y='Continuing_Writer_Count', hue='Procedure_Code', data=continuing_writers_count, marker='o')
plt.title('Number of Continuing Writers for Each Brand by Year')
plt.xlabel('Year')
plt.ylabel('Number of Continuing Writers')
plt.legend(title='Procedure Code (Brand)')

# Add data points as text
for line in ax.lines:
    for x_value, y_value in zip(line.get_xdata(), line.get_ydata()):
        ax.text(x_value, y_value, f'{y_value:.0f}', color=line.get_color(), ha='center', va='bottom')

x_axis= [2017, 2018]
ax.set_xticks(x_axis)
ax.set_xticklabels(x_axis)

plt.show()
```

**Observations:**
1. Fluctuating New Writer Trends: Varied annual trends in the number of new HCPs prescribing each brand, with some brands seeing increases and others decreases.
2. Overall Decline in Continuing Writers: Most brands are experiencing a decline in the number of continuing writers, particularly noticeable between 2017 and 2018.

**Insights:**
1. New Writer Acquisition: Gaining new HCPs who prescribe the product for the first time is critical for market growth and expansion.
2. Writer Retention Challenges: Keeping existing writers engaged is key to maintaining market share and ensuring sustained product performance.
3. Competitive Dynamics: Fluctuations in new and continuing writers may be indicative of the competitive forces at play within the market.
